```python
import xgboost as xgb
```

```python
result = pd.DataFrame()
params = {
        'eta': 0.1,
        'max_depth': 9,
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'seed': 0,
        'silent': 1
}
clos=[x for x in train.columns if x not in ['target', 'id']]  
xgbtrain = xgb.DMatrix(train[clos], train['target'])
xgbtest = xgb.DMatrix(test[clos])
watchlist = [(xgbtrain, 'train'), (xgbtrain, 'test')]
num_rounds = 150
model = xgb.train(params, xgbtrain, num_rounds, watchlist, early_stopping_rounds=50) 
test['target'] = model.predict(xgbtest, ntree_limit=model.best_ntree_limit)
r=test[['id','target']]
result=pd.concat([result,r])
result['id']=result['id'].astype('int')
result.to_csv("../result/result_v1.csv", index=False)
```
